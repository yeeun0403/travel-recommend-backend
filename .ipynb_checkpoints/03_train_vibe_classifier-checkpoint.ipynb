```python
# vibe 다중 라벨 분류기 코드 (Colab 기반, Drive 경로 고려)
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.metrics import classification_report
```

```python
# 1. 파일 경로 및 로딩
file_path = '/content/drive/MyDrive/졸업논문/강원도_관광지_20_예시.xlsx'
df = pd.read_excel(file_path)
df_vibe = df[['description', 'vibe']].dropna()
```

```python
# 2. 다중 라벨 처리
df_vibe['vibe'] = df_vibe['vibe'].apply(lambda x: [tag.strip() for tag in str(x).split(',')])
```

```python
# 3. 입력 및 출력
X = df_vibe['description']
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df_vibe['vibe'])
```

```python
# 4. 벡터화 + 분류기 정의
X_vec = TfidfVectorizer(max_features=1000).fit_transform(X)
clf = OneVsRestClassifier(LogisticRegression(max_iter=1000))
```

```python
# 5. 교차 검증 예측
kf = KFold(n_splits=5, shuffle=True, random_state=42)
y_pred = cross_val_predict(clf, X_vec, y, cv=kf)
```

```python
# 6. 평가 출력
report = classification_report(y, y_pred, target_names=mlb.classes_)
report
```

```python
import joblib

joblib.dump(clf, '/content/drive/MyDrive/졸업논문/vibe_model.pkl')
joblib.dump(mlb, '/content/drive/MyDrive/졸업논문/vibe_encoder.pkl')

print("✅ vibe 관련 모델, 인코더 저장 완료!")
```
